```python
import  numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu=100
sigma=5

data=np.random.normal(mu,sigma,10000)

x=np.linspace(mu -3*sigma,mu+3*sigma, 10000)
pdf=norm.pdf(x,mu,sigma)

plt.figure(figsize=(10,6))
plt.hist(data,bins=30, density=True,alpha=0.6,color='#1f77b4',edgecolor='black',linestyle='dashed',linewidth=1.5)
plt.plot(x,pdf,'r-',lw=2)
plt.title('Norml Distribution: $\mu$=100,$\sigma=5')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
x = np.linspace(-4, 4, 10000)
y = norm.pdf(x)
x1, x2 = -1, 1
plt.plot(x, y, label='Normal Distribution', color='blue')
plt.fill_between(x, y, where=(x > x1) & (x < x2), color='black', alpha=0.5, label='p(x1 < X < x2)')
plt.fill_between(x, y, where=(x > x1), color='black', alpha=0.3, label='p(X > x1)')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution with Shaded Areas')
plt.legend()
plt.grid(True)
plt.show()
```
