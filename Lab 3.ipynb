```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

N = 500
children_counts = np.random.choice([0, 1, 2, 3, 4], size=N, p=[0.1, 0.25, 0.35, 0.2, 0.1])

unique, counts = np.unique(children_counts, return_counts=True)
pmf = counts / N
cdf = np.cumsum(pmf)

plt.figure(figsize=(10, 6))
sns.barplot(x=unique, y=pmf, color='skyblue', edgecolor='black')
plt.xlabel("Number of children per family")
plt.ylabel("Probability (pmf)")
plt.title("Probability Mass Function (pmf) of Family Sizes")
plt.xticks(unique)
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x=unique, y=cdf, marker="o", color="red")
plt.xlabel("Number of children per family")
plt.ylabel("Cumulative probability (cdf)")
plt.title("Cumulative Distribution (cdf) of Family Sizes")
plt.xticks(unique)
plt.ylim(0, 1.1)
plt.show()
```
