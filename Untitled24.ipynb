```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

def plot_histogram(df, column, bins=10):
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], bins=bins, kde=True, color='blue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

def plot_scatter(df, x_column, y_column):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[x_column], y=df[y_column], color='red')
    plt.title(f'Scatter plot of {x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()

def plot_bar_chart(df, x_column, y_column):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=df[x_column], y=df[y_column], palette='viridis')
    plt.title(f'Bar chart of {y_column} by {x_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.show()

# Load the data
file_path = "/content/cricket_dataset-1.csv"
data = load_data(file_path)

# Plot histograms
plot_histogram(data, 'Runs')
plot_histogram(data, 'Wickets')

# Plot scatter plots
plot_scatter(data, 'Matches', 'Runs')
plot_scatter(data, 'Economy Rate', 'Wickets')

# Plot bar charts
plot_bar_chart(data, 'Player', '50s')
plot_bar_chart(data, 'Player', '100s')
```
